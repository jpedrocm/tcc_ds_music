```python
import pandas as pd
import seaborn as sns
```

```python
from tqdm import tqdm
```

```python
from ast import literal_eval
```

```python
tqdm.pandas()
```

```python
tcc = pd.read_csv("data/MUSIC4ALL/processed/tcc_music4all.csv", index_col=0)
```

```python
tcc.head()
```

```python
tcc.info()
```

```python
tcc.genres = tcc.genres.progress_apply(literal_eval)
```

```python
tcc.describe()
```

```python
tcc["explicit_cat"] = tcc.explicit.astype("category")
```

```python
tcc["mode_cat"] = tcc["mode"].astype("category")
```

```python
tcc["time_signature_cat"] = tcc.time_signature.astype("category")
```

```python
tcc["key_cat"] = tcc.key.astype("category")
```

```python
sns.set_theme(style="white", palette='Set2')
```

```python
tcc.describe(include="category")
```

# Mode

```python
fig = sns.countplot(x="mode", data=tcc);
fig.set(xlabel="Mode", ylabel='Contagem', ylim=(0,4100))
fig.get_figure().savefig('results/contagem_mode.png', dpi=300)
```

```python
fig = sns.countplot(x="mode", hue="explicit_cat", data=tcc);
fig.set(xlabel="Mode", ylabel='Contagem', ylim=(0,4100))
fig.legend_.set_title('Explicit')
fig.get_figure().savefig('results/contagem_mode_explicit.png', dpi=300)
```

# Time Signature

```python
fig = sns.countplot(x="time_signature_cat", data=tcc);
fig.set(xlabel="Time Signature", ylabel='Contagem', ylim=(0,6000))
fig.get_figure().savefig('results/contagem_signature.png', dpi=300)
```

```python
fig = sns.countplot(x="time_signature_cat", hue="explicit_cat", data=tcc);
fig.set(xlabel="Time Signature", ylabel='Contagem', ylim=(0,6000))
fig.legend_.set_title('Explicit')
fig.get_figure().savefig('results/contagem_signature_explicit.png', dpi=300)
```

# Key

```python
fig = sns.countplot(x="key_cat", data=tcc);
fig.set(xlabel="Key", ylabel='Contagem', ylim=(0,870))
fig.get_figure().savefig('results/contagem_key.png', dpi=300)
```

```python
fig = sns.countplot(x="key_cat", hue="explicit_cat", data=tcc);
fig.set(xlabel="Key", ylabel='Contagem', ylim=(0,870))
fig.legend_.set_title('Explicit')
fig.get_figure().savefig('results/contagem_key_explicit.png', dpi=300)
```
