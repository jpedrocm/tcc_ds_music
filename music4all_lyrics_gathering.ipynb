```python
import pandas as pd
from tqdm import tqdm

import glob
from os.path import basename
```

```python
lyrics = pd.DataFrame(columns=['id', 'lyrics'])
```

```python
for txt_path in tqdm(glob.glob("data/MUSIC4ALL/lyrics/*.txt")):
    with open(txt_path, 'r') as f:
        content = f.read()
        id_nb = basename(txt_path).split('.')[0]
        lyrics.loc[len(lyrics)] = [id_nb, content]
```
