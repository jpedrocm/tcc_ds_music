```python
import pandas as pd
import numpy as np
```

```python
import time
import json
import glob
from tqdm import tqdm
```

```python
from pprint import pprint
```

```python
data = pd.read_csv(f'data/MUSIC4ALL/processed/filtered_music4all_v4_part1.csv', index_col=0)
data.head()
```

```python
feat_json = []

for json_path in tqdm(glob.glob("data/MUSIC4ALL/crawled/audio_feat_results*.json")):
    print(json_path)
    with open(json_path, 'r') as f:
        json_content = json.load(f)
        feat_json.extend(json_content)
```

```python
len(feat_json)
```

```python
len(data)
```

```python
pprint(feat_json[0])
```

```python
FEAT_COLS = ['acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'key', 'liveness',
       'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence']

for json_data in tqdm(feat_json):
    id_ = json_data['id']
    change_values = [json_data[key] for key in FEAT_COLS]
    data.loc[data.spotify_id == id_, FEAT_COLS] = change_values
```

```python
data.head()
```

```python
info_json = []

for json_path in tqdm(glob.glob("data/MUSIC4ALL/crawled/track_results*.json")):
    print(json_path)
    with open(json_path, 'r') as f:
        json_content = json.load(f)['tracks']
        info_json.extend(json_content)
```

```python
COLS_1 = ['album_name', 'song', 'artist', 'popularity', 'explicit', 'release', 'markets']
COLS_2 = [c +'_2' for c in COLS_1]

for col in COLS_2:
    data[col] = pd.NA
```

```python
for cur_item in tqdm(info_json):    
    id_ = cur_item['id']
    
    album_name = cur_item['album']['name']
    song = cur_item['name']
    artist_name = cur_item['artists'][0]['name']
    popularity = cur_item['popularity']
    explicit = cur_item['explicit']
    markets = ','.join(cur_item['available_markets'])
    release = cur_item['album']['release_date']
    
    data.loc[data.spotify_id == id_, COLS_2] = [album_name, song, artist_name, popularity,
                                                explicit, release, markets]
```

```python
for c in COLS_2:
    data[c[:-2]] = data[c]
```

```python
final_core_data = data.drop(columns=COLS_2)
```
