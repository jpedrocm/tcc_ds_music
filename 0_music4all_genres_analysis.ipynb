```python
import pandas as pd
```

# Genres

```python
genres = pd.read_csv("data/MUSIC4ALL/raw/id_genres.csv", sep='\t')
genres.head()
```

```python
genres.info()
```

```python
genres.genres.value_counts()
```

```python
genre_list = genres.genres.tolist()
all_genres = [l.lower().strip().split(',') for l in genre_list]
all_genres[:5]
```

```python
genre_count = {}

for l in all_genres:
    for g in l:
        if g not in genre_count:
            genre_count[g] = 1
        else:
            genre_count[g] += 1
```

```python
pd.DataFrame.from_dict(genre_count, orient='index').plot();
```
