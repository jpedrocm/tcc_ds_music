# Language Detection

```python
data = pd.read_csv('data/MUSIC4ALL/processed/filtered_music4all_v1.csv', index_col=0)
```

## Google Langdetect

```python
data["google_lang"] = data.lyrics.progress_apply(ld.detect)
```

## FastText Langdetect

```python
#!python -m spacy download pt_core_news_md
```

```python
fasttext_nlp = spacy.load('pt_core_news_md')
fasttext_nlp.add_pipe("language_detector")

def get_fasttext_lang(text):
    return fasttext_nlp(text)._.language
```

```python
data["fasttext_lang"] = data.lyrics.progress_apply(get_fasttext_lang)
```
