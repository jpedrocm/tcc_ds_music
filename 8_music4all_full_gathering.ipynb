```python
import pandas as pd
```

```python
core = pd.read_csv("data/MUSIC4ALL/processed/filtered_music4all_v5_part1.csv", index_col=0)
core.info()
```

```python
complement = pd.read_csv("data/MUSIC4ALL/processed/deduplicated_ids_fixed_feat.csv", index_col=0)
complement.info()
```

```python
all(core.columns == complement.columns)
```

# Core Treatment

```python
core.head()
```

```python
core.time_signature = core.time_signature.astype(int)
```

```python
core.explicit = core.explicit.astype(int)
```

```python
core.genres = core.genres.apply(lambda s: s.strip().lower().split(','))
```

```python
core["nb_genres"] = core.genres.apply(lambda l: len(l))
```

```python
core['year'] = core.release.apply(lambda s: s[:4])
```

```python
core.year = core.year.astype(int)
```

```python
core[['key', 'mode', 'valence', 'tempo']].head()
```

```python
core.key = core.key.astype(int)
```

```python
core['mode'] = core['mode'].astype(int)
```

```python
core.head()
```

```python
core[['key', 'mode', 'valence', 'tempo']].head()
```

```python
final_core = core.drop(columns="release")
```

```python
final_core.info()
```

# Complement Treatment

```python
complement.head()
```

```python
complement.time_signature = complement.time_signature.astype(int)
```

```python
complement.explicit = complement.explicit.astype(int)
```

```python
complement.genres = complement.genres.apply(lambda s: s.strip().lower().split(','))
```

```python
complement["nb_genres"] = complement.genres.apply(lambda l: len(l))
```

```python
complement['year'] = complement.release.apply(lambda s: s[:4])
```

```python
complement.year = complement.year.astype(int)
```

```python
complement.markets = complement.markets.apply(lambda s: s.strip()\
                                                         .replace('[', '')\
                                                         .replace(']', '')\
                                                         .replace(' ', '')\
                                                         .replace('\'', ''))
```

```python
complement[['key', 'mode', 'valence', 'tempo']].head()
```

```python
complement.key = complement.key.astype(int)
```

```python
complement['mode'] = complement['mode'].astype(int)
```

```python
complement.head()
```

```python
complement[['key', 'mode', 'valence', 'tempo']].head()
```

```python
final_complement = complement.drop(columns="release")
```

```python
final_complement.info()
```

# Merge

```python
final_music4all = pd.concat([final_core, final_complement])
```

```python
final_music4all = final_music4all.sort_index()
```

```python
final_music4all.markets = final_music4all.markets.fillna("UNK")
```

```python
final_music4all.info()
```

```python
final_music4all.to_csv("data/MUSIC4ALL/processed/treated_music4all_v1.csv", index=True)
```
