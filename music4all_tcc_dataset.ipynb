```python
data = pd.read_csv("data/MUSIC4ALL/processed/treated_music4all_v3.csv", index_col=0)
```

```python
data['genres'] = data.genres.progress_apply(literal_eval)
```

```python
data.info()
```

```python
data = data.drop(columns=["markets"])
```

```python
genres = []

for i,d in data.iterrows():
    genres.extend(d.genres)
```

```python
genre_count = Counter(genres)
```

```python
chosen_genre = set([t[0] for t in genre_count.most_common(20)])
chosen_genre
```

```python
true = 0

for i, d in data.iterrows():
    if len(set(d.genres).intersection(chosen_genre)) > 0:
        true+=1
        
print(true==len(data))
```
