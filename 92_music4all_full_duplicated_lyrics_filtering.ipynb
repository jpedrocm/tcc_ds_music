# Approximate Lyric Similarity

```python
lyrics_df = data.copy()[['lyrics']]
```

```python
#!python -m spacy download pt_core_news_md
```

```python
nlp = spacy.load('pt_core_news_md')
```

```python
lyrics_df["lyrics_obj"] = lyrics_df.lyrics.str.lower().progress_apply(nlp)
```

```python
matches = {}
```

```python
for i, row_i in enumerate(tqdm(lyrics_df.iterrows())):
    index_i, data_i = row_i

    for index_j, data_j in lyrics_df.iloc[i+1:].iterrows():
        
        score = data_i.lyrics_obj.similarity(data_j.lyrics_obj)
        
        if score >= 0.99:
            if index_i not in matches:
                matches[index_i] = {index_j: score}
            else:
                matches[index_i][index_j] = score

            if index_j not in matches:
                matches[index_j] = {index_i: score}
            else:
                matches[index_j][index_i] = score
```

```python
with open('data/MUSIC4ALL/processed/lyrics_matches.json', 'w') as f:
    json.dump(matches, f)
```

### Removing

```python
with open('data/MUSIC4ALL/processed/lyrics_matches.json', 'r') as f:
    matches = json.load(f)
```

```python
len(matches.keys())
```

```python
def choose_priority_item(idxs):
        
    if len(idxs) == 1:
        return idxs[0]
        
    chosen_idx = None
    i = 0
    j = 1

    while i < len(idxs) and j < len(idxs):

        if i == j:
            j+=1
            continue
                
        if data[data.index==int(idxs[i])].year.values[0] <= data[data.index==int(idxs[j])].year.values[0]:
            chosen_idx = idxs[i]
            j+=1
        else:
            chosen_idx = idxs[j]
            i+=1
    
    return chosen_idx
```

```python
remove_idxs = set()
```

```python
for key, values in tqdm(matches.items()):
    complement = [k for k in values.keys() if k not in remove_idxs]
    
    if key not in remove_idxs:
        possibles = [key, *complement]
    else:
        possibles = [*complement]
    
    if len(possibles) > 0:
        chosen_idx = choose_priority_item(possibles)
        possibles.remove(chosen_idx)
        remove_idxs.update(possibles)
```

```python
len(remove_idxs)
```

```python
int_idxs = [int(idx) for idx in remove_idxs]
```

```python
final_data = data.drop(labels=int_idxs)
```
